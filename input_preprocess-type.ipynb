```python
import numpy as np
import pandas as pd
```

```python
read_csv=pd.read_csv("50_Startups.csv")
```

```python
read_csv.head()
```

```python
read=pd.get_dummies(read_csv,drop_first=True).astype(int)
```

```python
read.head()
```

```python
read.columns
```

```python
ind=read[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dep=read[['Profit']]
```

```python
# preprocess
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ind,dep,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# model creation:
from sklearn.svm import SVR
reg=SVR(C= 1000,gamma= 'scale', kernel= 'rbf')
reg.fit(x_train,y_train)
```

```python
# predict and evaluation:
from sklearn.metrics import r2_score
pred=reg.predict(x_test)
```

```python
score=r2_score(y_test,pred)
score
if score<0.50:
    print(f"{score} -- Very worst model...")
else:
    print("good model")
```

```python
# goes to deployement:
import pickle
file_name='single-pre_input_svr.sav'
create=pickle.dump(reg,open(file_name,'wb'))
```
